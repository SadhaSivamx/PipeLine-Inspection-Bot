### Sequence of Operations
1. GrayScale - Cracks are better identified in Grayscale [Grayscale Tutorial ](https://www.geeksforgeeks.org/python-grayscaling-of-images-using-opencv/)

2. Threshold - To make the hole part and normal part differentiatable [Threshold Tutorial](https://www.geeksforgeeks.org/image-thresholding-in-python-opencv/)

3. Erode - to kill the minor details and hold on to main ones [Erode Tutorial](https://www.geeksforgeeks.org/python-opencv-cv2-erode-method/)

4. Canny - to find the corners alone [Canny Algorithm](https://www.geeksforgeeks.org/implement-canny-edge-detector-in-python-using-opencv/)

5. Contors - Draw bounding lines for the cracks found [Contors Tutorial ](https://www.geeksforgeeks.org/find-and-draw-contours-using-opencv-python/)

```python
#GrayScale Img
import cv2 as cv
import matplotlib.pyplot as plt
imgo=cv.imread("Crack.jpg")
img=cv.cvtColor(imgo,cv.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
```

```python
#Thresholding Img
ret,Timg=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
plt.imshow(Timg,cmap="gray")
```

```python
#Eroding Image Features
import numpy as np
kernel=np.ones((5,5),int)
Erode=cv.erode(Timg, kernel, iterations=1)
plt.imshow(Erode,cmap="gray")
```

```python
Erode=cv.Canny(Erode, 30, 30)
contors, _ = cv.findContours(Erode, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cimg = cv.drawContours(imgo, contors, -1, (255, 0, 0), 2)
plt.imshow(cimg)
```

```python
def ShowCracks(img):
  import cv2 as cv
  import matplotlib.pyplot as plt
  imgo=cv.imread(img)
  img=cv.cvtColor(imgo,cv.COLOR_BGR2GRAY)
  #plt.imshow(img,cmap="gray")
  ret,Timg=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
  #plt.imshow(Timg,cmap="gray")
  import numpy as np
  kernel=np.ones((3,3),int)
  Erode=cv.erode(Timg, kernel, iterations=1)
  Erode=cv.Canny(Erode, 30, 30)
  contors, _ = cv.findContours(Erode, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
  cimg = cv.drawContours(imgo, contors, -1, (255, 0, 0), 2)
  plt.imshow(cimg)
```

```python
ShowCracks("Wall.jpeg")
```

```python
ShowCracks("crack.jpeg")
```

```python
ShowCracks("cracks.jpg")
```
